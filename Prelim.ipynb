```python
import pandas as pd
from google.colab import files

# Upload CSV
uploaded = files.upload()
filename = next(iter(uploaded))
df = pd.read_csv(filename)

print("SUS Dataset")
print(df)

# SUS Question Columns
sus_columns = [f"Q{i}" for i in range (1, 11)]

# Calculate SUS score
def calculate_sus_score(row):
  score = 0

  # Odd or Even Items
  for i, col in enumerate(sus_columns,start=1):
      response = row[col]

      # Odd item
      if i % 2 == 1:
        score += response -1

      # Even item
      else:
        score += 5 - response
  return score * 2.5

  # Display computed SUS Scores
df["sus_score"] = df.apply(calculate_sus_score,axis=1)
print("\individual SUS Scores")

if "user" in df.columns:
   print(df[["user", "sus_score"]])
else:
  print(df[["sus_score"]])

# Descriptive Statistics
mean_sus = df["sus_score"].mean()
median_sus = df["sus_score"].median()
var_sus = df["sus_score"].var()
std_sus = df["sus_score"].std()
cv_sus = (std_sus/mean_sus)*100

print("\nDescriptive Statistics")
print(f"Mean {mean_sus:.3f}")
print(f"Median {median_sus:.3f}")
print(f"Variance {var_sus:.3f}")
print(f"Standard Deviation {std_sus:.3f}")
print(f"Coefficient of Variation {cv_sus:.3f}")

# SUS Interpretation
if mean_sus >= 80.3:
  adj_rating="Excellent"
elif mean_sus >= 68.1:
  adj_rating="Good"
elif mean_sus >= 68:
  adj_rating="Okay"
elif mean_sus >= 51:
  adj_rating="Poor"
else:
  adj_rating="Awful"
print(f"Adjusted Rating: {adj_rating}")
```
